#Sử dụng bộ dữ liệu mtcars trong R vẽ các biểu đồ trực quan hóa độ lớn sau:

#- Biểu đồ barplot

```r
data(mtcars)
head(mtcars)
```

```r
# Đếm số lượng xe theo số xi lanh
cyl_counts <- table(mtcars$cyl)

# Vẽ biểu đồ barplot
barplot(cyl_counts,
        main = "So luong xe theo xi lanh",
        xlab = "So xi lanh",
        ylab = "So luong xe",
        col = c("skyblue", "orange", "forestgreen"))
```

```r
bar <- barplot(cyl_counts,
               main = "So luong xe theo so xi lanh",
               xlab = "So xi lanh",
               ylab = "So luong xe",
               col = "steelblue",
               ylim = c(0, max(cyl_counts) + 2))

# Thêm giá trị số trên đầu cột
text(x = bar, y = cyl_counts, label = cyl_counts, pos = 3, cex = 1.2, col = "black")
```

#- Biểu đồ barplot xếp chồng

```r
# Tạo bảng tần suất: số xe theo số xi lanh và số tốc độ
counts <- table(mtcars$cyl, mtcars$gear)

# Vẽ barplot xếp chồng
barplot(counts,
        main = "So luong xe theo so xi lanh va so toc do (gear)",
        xlab = "So xi lanh",
        ylab = "So luong xe",
        col = c("red", "blue", "green"),
        legend = rownames(counts),
        args.legend = list(title = "So toc do", x = "topright"))
```

#- Biểu đồ điểm

```r
plot(mtcars$cyl, mtcars$mpg,
     main = "Bieu do diem: mpg theo so xi lanh",
     xlab = "So xi lanh",
     ylab = "Miles Per Gallon (mpg)",
     pch = 19,            # Kiểu điểm
     col = "blue")
```

#- Biểu đồ nhiệt

```r
# Tính ma trận tương quan giữa các biến số trong mtcars
cor_matrix <- cor(mtcars)

# Vẽ heatmap
heatmap(cor_matrix,
        main = "Heatmap ma tran tuong quan cac bien trong mtcars",
        col = colorRampPalette(c("blue", "white", "red"))(20),
        symm = TRUE)
```
